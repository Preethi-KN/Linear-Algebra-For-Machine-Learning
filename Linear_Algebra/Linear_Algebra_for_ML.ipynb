```python
import numpy as np
V1=np.array([4,2,5,6,7])
dist=np.linalg.norm(V1)
print("Distence from the orign:",dist)
```

### Eucliden Distence

```python
V1=np.array([4,2,5,6,7])
V2=np.array([1,3,2,7,-3])
diff=V1-V2

dist=np.linalg.norm(diff)
#diff
print("Eucliden Distence between 2 vectors:",dist)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.distance import euclidean

# 1. Generate 5 random vectors
vectors = np.random.rand(5, 3)

# 2. Assign a random class  to each vector
classes = np.random.randint(0, 2, 5)

# 3. Plot the vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

for vec, cls in zip(vectors, classes):
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2], color=("r" if cls == 1 else "b"))


# 4. Get user input for a query point 
query_vector = np.array(list(map(float, input("Enter the query separated by space: ").split())))

# 5. Calculate the distance from the query vector to the 5 vectors and find the nearest neighbor
distances = [euclidean(query_vector, vec) for vec in vectors]
nearest_neighbor_index = np.argmin(distances)

# 6. Output the class of the nearest neighbor
print("The class of the nearest neighbor is:", classes[nearest_neighbor_index])

# 7. Plot the query vector with a different color-->green
ax.quiver(0, 0, 0, query_vector[0], query_vector[1], query_vector[2], color="g")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```

### Addition, Subtraction and mean centering

```python
#Creating the random 2D vector
data=np.random.rand(100,2)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.scatter(data[:, 0], data[:, 1], label="Original Data")
plt.xlabel("X-axis")
plt.ylabel("Y-Axis")
plt.title("Data before mean centring")
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.legend()
```

```python
#mean center pointing
data_mean=np.mean(data)
print(data_mean)

#subtract the mean from the data
mean_center_data=np.subtract(data,data_mean)

#now plot the mean center scatter plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.scatter(mean_center_data[:, 0], mean_center_data[:, 1], label="Mean center data")
#plt.scatter(mean_center_data[:, 0], mean_center_data[:, 1], label="Mean center data")
plt.ylabel("Y-Axis")
plt.title("Data after mean centring")
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.legend()
```
